Next, we can apply the gradient descent algorithm to the problem.

First, we need a function that calculates the derivative for the objective function.

```python
from numpy import asarray
from numpy.random import rand
from numpy.random import seed

def derv(x):
  return 2.0*x
def obj(x):
  return x**2.0
solutions,scores=list(),list()

def grad_des(obj,derv,bound,iter,lr):
  solution=bound[:, 0] +rand(len(bound)) * (bound[:, 1] - bound[:, 0])
  for i in range(iter):
    grad=derv(solution)
    solution=solution-lr*grad
    soln_eval=obj(solution)
    solutions.append(solution)
    scores.append(soln_eval)
    print(f'Run No {i} | f({solution[0]:.7f}) = {soln_eval[0]:.7f}')
  return[solution,soln_eval]


seed(4)
bound=asarray([[-1.0,1.0]])
iter=30
lr=0.1
best,score=grad_des(obj,derv,bound,iter,lr)
print(f'Best : {best} | Score : {score[0]:.7f}')
```

Graphical SGD

```python

from numpy import arange

inputs=arange(bound[0,0],bound[0,1]+0.1,0.1)
res=obj(inputs)
plt.plot(inputs,res)
plt.plot(solutions,scores,'.-',color='r')
```

SGD with momentum

```python
from numpy import asarray
from numpy.random import rand
from numpy.random import seed

def derv(x):
  return 2.0*x
def obj(x):
  return x**2.0
solutionsm,scoresm=list(),list()

def grad_des_momentum(obj,derv,bound,iter,lr,momentum):
  solution=bound[:, 0] +rand(len(bound)) * (bound[:, 1] - bound[:, 0])
  change = 0.0
  for i in range(iter):
    grad=derv(solution)
    newchange=lr*grad+momentum*change
    solution=solution-newchange
    change=newchange
    soln_eval=obj(solution)
    solutionsm.append(solution)
    scoresm.append(soln_eval)
    print(f'Run No {i} | f({solution[0]:.9f}) = {soln_eval[0]:.14f}')
  return[solution,soln_eval]

seed(4)
bound=asarray([[-1.0,1.0]])
iter=30
lr=0.1
momentum=0.3
best,score=grad_des_momentum(obj,derv,bound,iter,lr,momentum)
print(f'Best : {best[0]:.9f} | Score : {score[0]:.14f}')
```

Graphical form of SGD with momentum

```python
from numpy import arange

inputs=arange(bound[0,0],bound[0,1]+0.1,0.1)
res=obj(inputs)
plt.plot(inputs,res)
plt.plot(solutionsm,scoresm,'.-',color='r')
```
